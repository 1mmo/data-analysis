```python
import numpy as np

from sktime.datasets import load_airline
from sktime.forecasting.theta import ThetaForecaster

y = load_airline()
fh = np.arange(1, 13)

forecaster = ThetaForecaster(sp=12)
forecaster.fit(y, fh=fh)
```

```python
coverage = 0.90
y_pred_ints = forecaster.predict_interval(coverage=coverage)
y_pred_ints
```

```python
y_pred = forecaster.predict()

fig, ax = plot_series(
    y, y_pred, labels=['y', 'y_pred'], pred_interval=y_pred_ints
)
```

```python
import numpy as np
import pandas as pd
from sktime.datasets import load_airline
from sktime.utils.plotting import plot_series
```

```python
from sktime.forecasting.base import ForecastingHorizon
```

```python
fh = ForecastingHorizon(
    pd.PeriodIndex(pd.date_range('1961-01', periods=36, freq='M')),
    is_relative=False
)
```

```python
fh
```

```python
cutoff = pd.Period('1960-12', freq='M')
```

```python
fh.to_relative(cutoff)
```

```python
fh.to_absolute(cutoff)
```

```python
forecaster = NaiveForecaster(strategy='last')
```

```python
forecaster.fit(data)
```

```python
pred = forecaster.predict(fh)
```

```python
plot_series(data, pred, labels=['data', 'pred'])
```

```python
forecaster = NaiveForecaster(strategy='last', sp=12)
forecaster.fit(data)
pred = forecaster.predict(fh)
```

```python
from sktime.forecasting.ets import AutoETS
from sktime.forecasting.arima import AutoARIMA
```

```python
data = load_airline()
fh = np.arange(1, 37)
forecaster = AutoETS(auto=True, sp=12)
forecaster.fit(data)
pred = forecaster.predict(fh)
plot_series(data, pred, labels=['data', 'pred'])
```

```python
data = load_airline()
fh = np.arange(1, 37)
forecaster = AutoARIMA(sp=12)
forecaster.fit(data)
pred = forecaster.predict(fh)
plot_series(data, pred, labels=['data', 'pred'])
```

```python
from sktime.forecasting.model_selection import temporal_train_test_split
```

```python
plot_series(data)
```

```python
data_train, data_test = temporal_train_test_split(data, train_size=0.8)
```

```python
plot_series(data_train, data_test, labels=['train', 'test'])
```

```python
fh = data_test.index
```

```python
forecaster = NaiveForecaster(strategy='last', sp=12)
forecaster.fit(data_train)
pred = forecaster.predict(fh)
plot_series(data_train, data_test, pred, labels=['train', 'test', 'pred'])
```

```python
forecaster = AutoARIMA(sp=12)
forecaster.fit(data_train)
pred = forecaster.predict(fh)
plot_series(data_train, data_test, pred, labels=['data_train', 'data_test', 'pred'])
```

```python
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error, mean_absolute_error
```

```python
mean_absolute_percentage_error(data_test, pred, symmetric=False)
```

```python
from sktime.datasets import load_airline
from sktime.forecasting.ets import AutoETS
from sktime.utils.plotting import plot_series
```

```python
y = load_airline()
```

```python
# December 1957

y_1957Dec = y[:-36]

forecaster = AutoETS(auto=True, sp=12, n_jobs=-1)

fh = np.arange(1, 13)

forecaster.fit(y_1957Dec)

y_pred_1957Dec = forecaster.predict(fh)
```

```python
plot_series(y_1957Dec, y_pred_1957Dec, labels=['y_1957Dec', 'y_pred_1957Dec'])
```

```python
y[[-36]]
```

```python
# January 1958

y_1958Jan = y[[-36]]
forecaster.update(y_1958Jan)

y_pred_1958Jan = forecaster.predict(fh)
y_pred_1958Jan
```

```python
plot_series(
    y[:-35],
    y_pred_1957Dec,
    y_pred_1958Jan,
)
```

```python
# February 1958

y_1958Feb = y[[-35]]

forecaster.update(y_1958Feb, update_params=False)
```

```python
import matplotlib.pyplot as plt
from sktime.datasets import load_arrow_head
from sktime.datatypes import convert
```

```python
X, y = load_arrow_head(return_X_y=True)
X = convert(X, from_type='nested_univ', to_type='numpy3D')
```

```python
labels, counts = np.unique(y, return_counts=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
for label in labels:
    ax.plot(X[y == label, 0, :][0], label=label)
plt.legend()
```
